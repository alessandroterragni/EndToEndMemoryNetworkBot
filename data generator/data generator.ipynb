## Dataset Generator

```python
import random


# knowledge basics
main_focus = ('Attack', 'Defense', 'Tactics', 'Techniques', 'Passing')
age = ("Over 13", "Under 9", "Under 11", "Under 13")
phase = ("Warm-up", "Main", "Cool-down", "shot on goal")
group_size = ("Single player", "Group", "Team")
goalkeeper = ("0","zero", "1","one", "2","two", "3","three","4","Four",\
              "5","Five","6","six", "7","seven", "8","eight","9","nine")
```

```python
# knowledge basics questions
main_focus_question = "What should be the main focus of the training? Attack, Defense, Tactics, Techniques or Passing?$\n"
greetings_question = "hello!, what can I help you with today?$\n"
age_question = "What is the average age of the training group? Under 9, Under 11, Under 13 or Over13?$\n"
phase_question = "In which phase do you want to train? Warm-up, Main, Cool-down, shot on goal?$\n"
group_size_question = "Who you want to train: a single player, a Group or the whole Team?$\n"
goalkeeper_question = "How many goalkeepers will join the training?$\n"
```

```python
# randomly generate an api call from the knowledge basics
def generateApiCall():
    apiCall = []
    rMF = random.randint(0,len(main_focus)-1)
    rA = random.randint(0,len(age)-1)
    rP = random.randint(0,len(phase)-1)
    rGS = random.randint(0,len(group_size)-1)
    rG = random.randint(0,len(goalkeeper)-1)
    apiCall.append(main_focus[rMF])
    apiCall.append(age[rA])    
    apiCall.append(phase[rP])
    apiCall.append(group_size[rGS])
    apiCall.append(goalkeeper[rG])
    return apiCall
```

```python
# generate a greeting question to start the conversation
greetings = ("hello", "hi", "hey", "how are you?", "hello", "is everything fine?")

def generateGreetings():
    r = random.randint(0,len(greetings)-1)
    return greetings[r]
```

```python
# single intent phrases
mainFocusPhrases = ("{}","I want to train on {}","I want to improve the {} skills",\
                    "I would like to train on {}", "I want to focus on {}",\
                   "I want to exercise the {}", "I would like to exercise the {}")
agePhrases = ("The average age is {}","{}")
phasePhrases = ("During the {} phase", "I want to train during the {} phase", \
                "I would like to train during the {} phase", "{} phase", "{}")
groupSizePhrases = ("{}", "I would like to train the {}", "I would like to work with the {}")
goalkeepersPhrases = ("{}","just {}", "only {}")

# generate a single intent sentence
def generateSinglePhrase(theme, n):
    r = random.randint(0,len(theme)-1)
    api_element = generateApiCall()[n]
    phrase = ""
    phrase += theme[r].format(api_element)
    return phrase, api_element
```

```python
# double intent phrases
mainFocus_phasePhrases = ("I want to train on {apiElement1} during the {apiElement2} phase",\
                         "I would like to train on {apiElement1} during the {apiElement2}phase",\
                         "I want to exercise on {apiElement1} during the {apiElement2} phase",\
                         "I would like to exercise on {apiElement1} during the {apiElement2} phase",\
                         "I want to focus on {apiElement1} during the {apiElement2} phase")




mainFocus_groupSizePhrases = ("I want to train a {apiElement2} on {apiElement1}",\
                             "I would like to train a {apiElement2} on {apiElement1}",\
                             "I want to exercise a {apiElement2} on {apiElement1}",
                             "My {apiElement2} struggles in {apiElement1}",
                             "My {apiElement2} suffers in {apiElement1}")




mainFocus_goalKeeperPhrases = ("I want to train on {apiElement1} with {apiElement2} goal keepers",\
                       "I would like to train on {apiElement1} with {apiElement2} goal keepers",\
                       " I want to focus on {apiElement1} with {apiElement2} goal keepers",\
                       "I want to exercise on {apiElement1} with {apiElement2} goal keepers")


age_groupSizePhrases = ("I want to train my {apiElement1} {apiElement2}", "I would like to train my {apiElement1} {apiElement2}")


# generate a double intent sentence
def generateDoublePhrase(phraseSet, n1, n2):
    r = random.randint(0,len(phraseSet)-1)
    apiElement1 = generateApiCall()[n1]
    apiElement2 = generateApiCall()[n2]
    phrase = ""
    phrase += phraseSet[r].format(apiElement1 = apiElement1, apiElement2 = apiElement2)
    return phrase, apiElement1, apiElement2 
```

```python
# triple intent phrases
mainFocus_phase_goalkeeeperPhase = ("I want to train on {apiElement1} with {apiElement3} goalkeepers during the {apiElement2} phase",\
                        "I would like to train on {apiElement1} during the {apiElement2} phase with {apiElement3} goalkeepers",\
                       "I want to exercise on {apiElement1} during the {apiElement2} phase with {apiElement3} goalkeepers", \
                        "I would like to exercise on {apiElement1} with {apiElement3} goalkeepers during the {apiElement2} phase ")

mainFocus_age_groupsizePhrases = ("I want to train my {apiElement2} {apiElement3} on {apiElement1}",\
                        "I want to train on {apiElement1} my {apiElement2} {apiElement3}",\
                       "I want to improve the {apiElement1} skills of my {apiElement2} {apiElement3}",\
                        "I would like to train my {apiElement2} {apiElement3} on {apiElement1}",\
                        "My {apiElement2} {apiElement3} suffers in the {apiElement1} phase",\
                       "The {apiElement2} {apiElement3} struggles in the {apiElement1} phase",\
                       "I want to exercise my {apiElement2} {apiElement3} on {apiElement1}", \
                        "I would like to exercise my {apiElement2} {apiElement3} on {apiElement1}")


# generate triple intent sentence 
def generateTriplePhrase(phraseSet, n1, n2, n3):
    r = random.randint(0,len(phraseSet)-1)
    apiElement1 = generateApiCall()[n1]
    apiElement2 = generateApiCall()[n2]
    apiElement3 = generateApiCall()[n3]
    phrase = ""
    phrase += phraseSet[r].format(apiElement1 = apiElement1, apiElement2 = apiElement2, apiElement3 = apiElement3)
    return phrase, apiElement1, apiElement2, apiElement3 
```

```python
# quadruple intent phrases
mainFocus_age_groupsize_phasePhrases = ("I want to train my {apiElement2} {apiElement3} on {apiElement1} during the {apiElement4} phase",\
                        "I want to train on {apiElement1} my {apiElement2} {apiElement3} during the {apiElement4} phase",\
                       "I want to improve the {apiElement1} skills of my {apiElement2} {apiElement3} during the {apiElement4} phase",\
                        "I would like to train my {apiElement2} {apiElement3} on {apiElement1} during the {apiElement4} phase",\
                        "My {apiElement2} {apiElement3} suffers in the {apiElement1} phase, I want to train this aspect during the {apiElement4} phase",\
                       "The {apiElement2} {apiElement3} struggles in the {apiElement1} phase, I want to train this aspect during the {apiElement4} phase",\
                       "I want to exercise my {apiElement2} {apiElement3} on {apiElement1} during the {apiElement4} phase", \
                        "I would like to exercise my {apiElement2} {apiElement3} on {apiElement1} during the {apiElement4} phase")

mainFocus_age_groupsize_goalkeepersPhrases = ("I want to train my {apiElement2} {apiElement3} on {apiElement1} with {apiElement4} goalkeepers",\
                        "I want to train on {apiElement1} my {apiElement2} {apiElement3} with {apiElement4} goalkeepers",\
                       "I want to improve the {apiElement1} skills of my {apiElement2} {apiElement3} with {apiElement4} goalkeepers",\
                        "I would like to train my {apiElement2} {apiElement3} on {apiElement1} with {apiElement4} goalkeepers",\
                        "My {apiElement2} {apiElement3} suffers in the {apiElement1} phase, I want to train this aspect with {apiElement4} goalkeepers",\
                       "The {apiElement2} {apiElement3} struggles in the {apiElement1} phase, I want to train this aspect with {apiElement4} goalkeepers",\
                       "I want to exercise my {apiElement2} {apiElement3} on {apiElement1} with {apiElement4} goalkeepers", \
                        "I would like to exercise my {apiElement2} {apiElement3} on {apiElement1} with {apiElement4} goalkeepers")

# generate quadruple intent sentence
def generateQuadruplePhrase(phraseSet, n1, n2, n3, n4):
    r = random.randint(0,len(phraseSet)-1)
    apiElement1 = generateApiCall()[n1]
    apiElement2 = generateApiCall()[n2]
    apiElement3 = generateApiCall()[n3]
    apiElement4 = generateApiCall()[n4]
    phrase = ""
    phrase += phraseSet[r].format(apiElement1 = apiElement1, apiElement2 = apiElement2, apiElement3 = apiElement3, apiElement4 = apiElement4)
    return phrase, apiElement1, apiElement2, apiElement3, apiElement4 
```

```python
# build an api call candidate, given the elements of the api call
def build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement):
    api_call = ""
    api_call += "apicall:  " + "MAIN_FOCUS = " + mainFocusApiElement + "  " + "AGE = " + ageApiElement + "  " +\
    "PHASE = " + phaseApiElement + "  " + "GROUP_SIZE = " + groupSizeApiElement + " "+\
    "GOAL_KEEPERS = " + goalKeepersApiElement
    return api_call


# generate a full dialogue: change r to select which type of dialogue you want to generate
def generateDialog():
    dialog = ""
    r = random.randint(0,8)
    
    # single intent
    if(r==0):
        mainFocus, mainFocusApiElement = generateSinglePhrase(mainFocusPhrases,0)
        age, ageApiElement = generateSinglePhrase(agePhrases,1)
        phase, phaseApiElement = generateSinglePhrase(phasePhrases,2)
        groupSize, groupSizeApiElement = generateSinglePhrase(groupSizePhrases,3) 
        goalKeepers, goalKeepersApiElement =generateSinglePhrase(goalkeepersPhrases,4)
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
    
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus + '\t' + age_question +\
        "3 " + age + '\t' + phase_question +\
        "4 " + phase + '\t' + group_size_question +\
        "5 " + groupSize + '\t' + goalkeeper_question +\
        "6 " + goalKeepers + '\t' + api_call +"$\n\n"
        
    #double intent 
    if(r==1):
        mainFocus_phase, mainFocusApiElement, phaseApiElement = generateDoublePhrase(mainFocus_phasePhrases,0,2)
        age, ageApiElement = generateSinglePhrase(agePhrases,1)
        groupSize, groupSizeApiElement = generateSinglePhrase(groupSizePhrases,3) 
        goalKeepers, goalKeepersApiElement =generateSinglePhrase(goalkeepersPhrases,4)
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
        
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus_phase + '\t' + age_question +\
        "3 " + age + '\t' + group_size_question +\
        "4 " + groupSize + '\t' + goalkeeper_question +\
        "5 " + goalKeepers + '\t' + api_call +"$\n\n"
      
    if(r==2):
        mainFocus_phrase, mainFocusApiElement, groupSizeApiElement = generateDoublePhrase(mainFocus_groupSizePhrases,0,3)
        age, ageApiElement = generateSinglePhrase(agePhrases,1)
        phase, phaseApiElement = generateSinglePhrase(phasePhrases,2) 
        goalKeepers, goalKeepersApiElement =generateSinglePhrase(goalkeepersPhrases,4)
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
        
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus_phrase + '\t' + age_question +\
        "3 " + age + '\t' + phase_question +\
        "4 " + phase + '\t' + goalkeeper_question +\
        "5 " + goalKeepers + '\t' + api_call +"$\n\n"
    
    if(r==3):
        mainFocus_phrase, mainFocusApiElement, goalKeepersApiElement = generateDoublePhrase(mainFocus_goalKeeperPhrases,0,4)
        age, ageApiElement = generateSinglePhrase(agePhrases,1)
        phase, phaseApiElement = generateSinglePhrase(phasePhrases,2)
        groupSize, groupSizeApiElement = generateSinglePhrase(groupSizePhrases,3) 
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
    
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus_phrase + '\t' + age_question +\
        "3 " + age + '\t' + phase_question +\
        "4 " + phase + '\t' + group_size_question +\
        "5 " + groupSize + '\t' + api_call +"$\n\n"
        
    if(r==4):
        age_group_phrase, ageApiElement, groupSizeApiElement = generateDoublePhrase(age_groupSizePhrases,1,3)
        mainFocus, mainFocusApiElement = generateSinglePhrase(mainFocusPhrases,0)
        phase, phaseApiElement = generateSinglePhrase(phasePhrases,2)
        goalKeepers, goalKeepersApiElement =generateSinglePhrase(goalkeepersPhrases,4)
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
    
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + age_group_phrase + '\t' + main_focus_question +\
        "3 " + mainFocus + '\t' + phase_question +\
        "4 " + phase + '\t' + goalkeeper_question +\
        "5 " + goalKeepers + '\t' + api_call +"$\n\n"
        
    
    # triple intent
    if(r==5):
        mainFocus_phrase, mainFocusApiElement,\
        phaseApiElement, goalKeepersApiElement  = generateTriplePhrase(mainFocus_phase_goalkeeeperPhase,0,2,4)
        age, ageApiElement = generateSinglePhrase(agePhrases,1)
        groupSize, groupSizeApiElement = generateSinglePhrase(groupSizePhrases,3) 
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
    
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus_phrase + '\t' + age_question +\
        "3 " + age + '\t' + group_size_question +\
        "3 " + groupSize + '\t'+ api_call +"$\n\n"

    
    if(r==6):
        mainFocus_phrase, mainFocusApiElement,\
        ageApiElement, groupSizeApiElement = generateTriplePhrase(mainFocus_age_groupsizePhrases,0,1,3)
        phase, phaseApiElement = generateSinglePhrase(phasePhrases,2)
        goalKeepers, goalKeepersApiElement =generateSinglePhrase(goalkeepersPhrases,4)
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
    
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus_phrase + '\t'+ phase_question +\
        "3 " + phase + '\t' + goalkeeper_question +\
        "4 " + goalKeepers + '\t' + api_call + "$\n\n"   
    
    # quadruple intent 
    if(r==7):
        mainFocus_phrase, mainFocusApiElement,\
        ageApiElement,groupSizeApiElement,\
        phaseApiElement = generateQuadruplePhrase(mainFocus_age_groupsize_phasePhrases,0,1,3,2)
        goalKeepers, goalKeepersApiElement =generateSinglePhrase(goalkeepersPhrases,4)
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
        
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus_phrase + '\t'+ goalkeeper_question +\
        "3 " + goalKeepers + '\t' + api_call + "$\n\n"
        
    if(r==8):
        mainFocus_phrase, mainFocusApiElement,\
        ageApiElement,groupSizeApiElement,\
        goalKeepersApiElement = generateQuadruplePhrase(mainFocus_age_groupsize_goalkeepersPhrases,0,1,3,4)
        phase, phaseApiElement = generateSinglePhrase(phasePhrases,2)
        api_call = build_api_call(mainFocusApiElement,ageApiElement,phaseApiElement,groupSizeApiElement,goalKeepersApiElement)
        
        dialog += "1 " + generateGreetings() + '\t' + greetings_question + \
        "2 " + mainFocus_phrase + '\t'+ phase_question +\
        "3 " + phase + '\t' + api_call + "$\n\n"
        
    return dialog, api_call
```

```python
# candidates questions
c_main_focus_question = "1 What should be the main focus of the training? Attack, Defense, Tactics, Techniques or Passing?\n"
c_greetings_question = "1 hello!, what can I help you with today?\n"
c_age_question = "1 What is the average age of the training group? Under 9, Under 11, Under 13 or Over13?\n"
c_phase_question = "1 In which phase do you want to train? Warm-up, Main, Cool-down, shot on goal?\n"
c_group_size_question = "1 Who you want to train: a single player, a Group or the whole Team?\n"
c_goalkeeper_question = "1 How many goalkeepers will join the training?\n"

candidates = ""
candidates += c_greetings_question
candidates += c_main_focus_question
candidates += c_age_question
candidates += c_phase_question
candidates += c_group_size_question
candidates += c_goalkeeper_question

dialogs = ""

count = 0
# randomly generate 3000 different dialogues 
while( count <= 3000):
    d, c = generateDialog()
    if d not in dialogs:
        dialogs += d
        candidates += "1 " + c + "\n" 
        count= count+1

# divide the dialogues in test, train and val
split = dialogs.split("$")
test = "".join(split[0:1000])
train = "".join(split[1000:2000])
val = "".join(split[2000:3000])

# save the files 
with open('task1-API-calls-dev.txt', 'a') as the_file:
        the_file.write(val)
with open('task1-API-calls-trn.txt', 'a') as the_file:
        the_file.write(train)
with open('task1-API-calls-tst.txt', 'a') as the_file:
        the_file.write(test)

with open('candidates.txt', 'a') as the_file:
        the_file.write(candidates)
        
# be careful, the splitting can left the start or the end of the files with incompleted dialogues,
# always check them manually and, in case, manually eliminate them from the beginning\end
```
